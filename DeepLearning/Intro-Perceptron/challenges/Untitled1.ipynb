```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import Imputer,StandardScaler
```

```python
data=pd.read_csv("./datasets2/Train/Train_Data.csv")
```

```python
Y=data["SalePrice"]
X=data.drop(columns=["SalePrice"])
```

```python
X
```

```python
from pandas.api.types import is_integer_dtype,is_float_dtype
```

```python
col=X.columns.tolist()
```

```python
numeric_type=[]
string_type=[]

for c in col:
    if is_integer_dtype(X[c]) or is_float_dtype(X[c]):
        numeric_type.append(c)
    else:
        string_type.append(c)
```

```python
print(len(numeric_type),len(string_type))
```

```python
num_df=X[numeric_type]
str_df=X[string_type]
```

```python
imp1=Imputer(missing_values='NaN',strategy='mean',axis=0)
```

```python
num_df=imp1.fit_transform(num_df)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp2=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
```

```python
str_df=imp2.fit_transform(str_df)
```

```python
str_df=pd.DataFrame(str_df)
```
