```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Xdata=pd.read_csv("./datasets/Training Data/Logistic_X_Train.csv").values
```

```python
Ydata=pd.read_csv("./datasets/Training Data/Logistic_Y_Train.csv").values
```

```python
Ydata=Ydata.reshape(-1)
```

```python
print(Ydata)
```

```python
plt.scatter(Xdata[:,0],Xdata[:,1],c=Ydata)
plt.show()
```

```python
print(Xdata.shape)
```

```python
from keras import models
from keras.layers import Dense
```

```python
model=models.Sequential()
model.add(Dense(16,activation='relu',input_shape=Xdata[0].shape))
model.add(Dense(16,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='Adadelta',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
hist=model.fit(Xdata,Ydata,epochs=500,batch_size=128,validation_split=0.2)
```

```python
h=hist.history
```

```python
plt.style.use('seaborn')
plt.plot(h['val_loss'], label='Validation loss')
plt.plot(h['loss'],label='Training loss')
plt.xlabel='epochs'
plt.ylabel='loss'
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn')
plt.plot(h['val_accuracy'], label='Validation accuracy')
plt.plot(h['accuracy'],label='Training accuracy')
plt.xlabel='epochs'
plt.ylabel='acc'
plt.legend()
plt.show()
```

```python
model.evaluate(Xdata,Ydata)[1]
```
