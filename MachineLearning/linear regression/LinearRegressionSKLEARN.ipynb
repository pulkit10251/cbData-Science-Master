```python
# SKLEARN - LINEAR REGRESSION
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=pd.read_csv('./Training Data/Linear_X_Train.csv').values
Y=pd.read_csv('./Training Data/Linear_Y_Train.csv').values
```

```python
Y.shape
```

```python
plt.style.use("seaborn")
plt.scatter(X,Y)
plt.show()
```

```python
model=LinearRegression()
```

```python
model.fit(X,Y)
```

```python
Xtest=pd.read_csv("./Test Cases/Linear_X_Test.csv").values
```

```python
ypred=model.predict(Xtest)
```

```python
df=pd.DataFrame(data=ypred,columns=["y"])
df.to_csv('submissiona.csv',index=False)
```

```python
intercept=model.intercept_
slope=model.coef_
print(intercept)
print(slope)
```

```python
y=slope[0][0]*X+intercept
```

```python
plt.scatter(X,Y)
plt.plot(X,y,color="orange")
```
