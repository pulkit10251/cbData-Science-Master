```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
```

```python
a=[1,2,3]
b=[4,5,6,7]

a,b=np.meshgrid(a,b)
print(a)
print(b)
```

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.plot_surface(a,b,a+b,cmap="rainbow")
plt.show()
```

```python
a=np.arange(-1,1,0.01)
b=a
a,b=np.meshgrid(a,b)

fig=plt.figure()
axes=fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap="rainbow")
plt.show()
```

```python
### contour plots ###

fig=plt.figure()
axes=fig.gca(projection='3d')
axes.contour(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```
