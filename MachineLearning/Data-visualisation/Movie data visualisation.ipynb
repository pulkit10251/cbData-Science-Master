# movie data visualisation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("movie_metadata.csv")

```

```python
df.head(10)
```

```python
df.columns
```

```python
titles=list(df.get('movie_title'))
```

```python
print(titles[:10])
```

```python
freq_titles={}

for title in titles:
    length=len(title)
    if freq_titles.get(length) is None:
        freq_titles[length]=1
    else:
        freq_titles[length]+=1
        
        
```

```python
print(freq_titles)
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
print(X,Y)
```

```python
plt.scatter(X,Y)
plt.xlabel("length of words")
plt.ylabel("frequency of occurence")
plt.title("movie data visualisation")
plt.show()
```
