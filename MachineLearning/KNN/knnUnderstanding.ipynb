```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

print(type(X))
qpoin=[2,3]
m=X.shape[0];
vals=[]
for i in range(m):
    x=dist(qpoin,X[i])
    vals.append((x,Y[i]))
```

```python

print(vals)
```

```python
vals=sorted(vals)
print(vals)
```

```python
vals=vals[:20]
print(vals)
```

```python
vals=np.array(vals)
print(vals)
print(vals[:,1])
```
