```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X,Y=make_regression(n_samples=400,n_features=1,noise=1.8,n_informative=1,random_state=11)

print(X.shape)
#print(Y.shape)
Y=Y.reshape((-1,1))
print(Y.shape)
```

```python
plt.scatter(X,Y)
```

```python
# normalize 
u=np.mean(X)
std=np.std(X)
X=(X-u)/std
```

```python
plt.style.use("seaborn")
plt.scatter(X,Y)
plt.show()
```

```python
ones=np.ones((X.shape[0],1))
X_=np.hstack((ones,X))
```

```python
print(X_)
```

```python
def predict(X,theta):
    return np.dot(X,theta)

def getThetaClosedForm(X,Y):
    Y=np.mat(Y)
    firstpart=np.dot(X.T,X)
    secondpart=np.dot(X.T,Y)
    
    theta=np.linalg.pinv(firstpart)*secondpart
    
    return theta
```

```python
theta=getThetaClosedForm(X_,Y)
print(theta)
```

```python
Y_=predict(X_,theta)
```

```python
plt.scatter(X,Y)
plt.plot(X,Y_,color='red',label='prediction')
plt.legend()
plt.show()
```
