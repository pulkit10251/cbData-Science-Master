```python
def hangman(count):
    list=["superman", "batman", "flash","aquaman","deadpool","venom","hackeye","blackwidow","hulk","thor","antman","ironman","captain america"]
    return list[count]

def generate(word):
    str=""
    for i in range(len(word)):
        str=str+"_ "
        
    return str
    
```

```python
def mistakes(chance):
    switcher={
        9: "-------\n",
        8: "-------\n"
           "  0    \n",
        7: "-------\n"
           "  0    \n"
           "  |    \n",
        6: "-------\n"
           "  0    \n"
           "  |    \n"
           " /     \n",
        5: "-------\n"
           "  0    \n"
           "  |    \n"
           " / \   \n",
        4: "-------\n"
           "\ 0    \n"
           "  |    \n"
           " / \   \n",
        3: "-------\n"
           "\ 0 /  \n"
           "  |    \n"
           " / \   \n",
        2: "-------\n"
           "\ 0 / |\n"
           "  |    \n"
           " / \   \n",
        1: "-------\n"
           "\ 0 /_|\n"
           "  |    \n"
           " / \   \n",
        0: "-------\n"
           "  0_|  \n"
           " /|\   \n"
           " / \   \n",
        
    }
    return switcher.get(chance)
```
