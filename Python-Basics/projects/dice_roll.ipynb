```python
def dicegen(b):
    a=[["   ","   ","   "],["   ","   ","   "],["   ","   ","   "]]
    if b==1 :
        a[1][1]=" O "
    elif b==2:
        a[0][1]=" O "
        a[2][1]=" O "
    elif b==3:
        a[0][1]=" O "
        a[1][1]=" O "
        a[2][1]=" O "
    elif b==4:
        a[0][0]=" O "
        a[2][0]=" O "
        a[0][2]=" O "
        a[2][2]=" O "
    elif b==5:
        a[0][0]=" O "
        a[2][0]=" O "
        a[0][2]=" O "
        a[2][2]=" O "
        a[1][1]=" O "
    else:
        a[0][0]=" O "
        a[1][0]=" O "
        a[2][0]=" O "
        a[0][2]=" O "
        a[1][2]=" O "
        a[2][2]=" O "
    
    return a
```
